```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
filename = "day25.txt"

with open(filename) as f:
    for line in f:
        nodes=[x.strip() for x in line.split(':')]
        dest=nodes[1].split()
        #G.add_node(dest)
        print(dest)
        for d in dest:
            G.add_edge(nodes[0],d)
            
```

```python
Gprime=G.remove_edges_from(nx.minimum_edge_cut(G))
```
