```python
def can_be_placed(fullstring, location, length):
    if location + length>len(fullstring):
        return False
    elif '.' in fullstring[location:location+length]:
        return False
    elif location != 0 and fullstring[location-1]=='#':
        return False
    elif location+length < len(fullstring) and fullstring[location+length:location+length+1]=='#':
        return False
    else:
        return True
        
    
    
    
    
```
